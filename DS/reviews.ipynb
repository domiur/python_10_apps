```python
import pandas
import matplotlib.pyplot as plt 
```

```python
df=pandas.read_csv("reviews.csv",parse_dates=["Timestamp"])
```

```python
df['Day']=df["Timestamp"].dt.date
df_avg=df.groupby('Day').mean()
```

```python
df
```

```python
df_avg
```

```python
plt.figure(figsize=(25,4))
plt.plot(df_avg.index,df_avg['Rating'])
```

```python
df['Week']=df["Timestamp"].dt.strftime("%Y-%U")
df_avg=df.groupby('Week').mean()
plt.figure(figsize=(25,4))
plt.plot(df_avg.index,df_avg['Rating'])
```

```python
df['Month']=df["Timestamp"].dt.strftime("%Y-%m")
df_avg=df.groupby('Month').mean()
plt.figure(figsize=(25,4))
plt.plot(df_avg.index,df_avg['Rating'])
```

```python
df['Month']=df["Timestamp"].dt.strftime("%Y-%m")
df_avg=df.groupby(['Month','Course Name']).mean().unstack()
df_avg.head()
df_avg.plot(figsize=(25,8))
```

```python
df_avg.head()
```

```python
df['dow']=df["Timestamp"].dt.strftime("%A")
df['dow_number']=df["Timestamp"].dt.strftime("%w")

df_avg=df.groupby(['dow','dow_number']).mean().sort_values('dow_number')
print(df_avg.head())
plt.figure(figsize=(25,8))
plt.plot(df_avg.index.get_level_values(0),df_avg['Rating'])
```
